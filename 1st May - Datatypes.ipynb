Data Types:- A data type defines the kind of value a variable can hold, python has several built-in data types...

```python
#01 int:- integer values(whole numbers)

x = 10
print(type(x)) # <class 'int'>
```

```python
#02 float:- floating point values(decimal numbers)

PI = 3.14
print(type(PI)) # <class 'float'>
```

```python
#03 str:- strings(sequence of characters like words & sentences)

name = "Mohammad Sufiyan"
print(type(name)) # <class 'str'>
```

```python
#04 bool:- boolean values(True / False)

isTeacher = True
print(type(isTeacher)) # <class 'bool'>
```

```python
#05 None:- integer values(whole numbers)

empty_var = None
print(type(empty_var)) # <class 'NoneType'>
```
