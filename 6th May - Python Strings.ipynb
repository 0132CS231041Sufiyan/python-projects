# PYTHON STRINGS
* Strings in python are surrounded by either single quotation marks, or double quotation marks,
  like:-'hello' is the same as "hello"...

```python
#01. Example:- You can display a string literal with the print() function:

print("Sufiyan")
print('Sufiyan')
```

```python
#02. Quotes Inside Quotes:- You can use quotes inside a string, as long as they don't match the quotes surrounding the string.

print("It's alright")
print("He is called 'Sufiyan'")
print('He is called "Sufiyan"')
```

```python
#03. Assign String to a Variable:- Assigning a string to a variable is done with the variable name followed by an equal sign and the string.

a = "Sufiyan"
print(a)
```

```python
#04. Multiline Strings:- You can assign a multiline string to a variable by using three quotes.

a = """Lorem ipsum dolor sit amet,
consectetur adipiscing elit,
sed do eiusmod tempor incididunt
ut labore et dolore magna aliqua."""
print(a)
```

```python
#05. Strings are Arrays:- Get the character at position 1 (remember that the first character has the position 0).

a = "Hello, World!"
print(a[1])
```

```python
#06. Looping Through a String:- Loop through the letters in the word "banana".

for x in "banana":
  print(x)
```

```python
#07. String Length:- The len() function returns the length of a string.

a = "Hello, World!"
print(len(a))
```

# Python - Slicing Strings
* You can return a range of characters by using the slice syntax.
* Specify the start index and the end index, separated by a colon, to return a part of the string

```python
#01. Example:- Get the characters from position 2 to position 5 (not included).

b = "Hello, World!"
print(b[2:5])
```

```python
#02. Slice From the Start:- By leaving out the start index, the range will start at the first character.

b = "Hello, World!"
print(b[:5]) #Get the characters from the start to position 5 (not included).
```

```python
#03. Slice To the End:- By leaving out the end index, the range will go to the end:

b = "Hello, World!"
print(b[2:]) #Get the characters from position 2, and all the way to the end.
```

```python
#04. Negative Indexing:- Use negative indexes to start the slice from the end of the string.

b = "Hello, World!"
print(b[-5:-2]) #From: "o" in "World!" (position -5) To, but not included: "d" in "World!" (position -2):
```

# Python - Modify Strings
* Python has a set of built-in methods that you can use on strings...

```python
#01. Upper Case:- The upper() method returns the string in upper case.

a = "Hello, World!"
print(a.upper())
```

```python
#02. Lower Case:- The lower() method returns the string in lower case.

a = "Hello, World!"
print(a.lower())
```

```python
#03. Remove Whitespace:- Whitespace is the space before and/or after the actual text, and very often you want to remove this space.

#Example:- The strip() method removes any whitespace from the beginning or the end.

a = " Hello, World! "
print(a.strip()) # returns "Hello, World!"
```

```python
#04. Replace String:- The replace() method replaces a string with another string.

a = "Hello, World!"
print(a.replace("H", "J"))
```

```python
#05. Split String:- The split() method returns a list where the text between the specified separator becomes the list items.

#Example:-The split() method splits the string into substrings if it finds instances of the separator.

a = "Hello, World!"
print(a.split(",")) # returns ['Hello', ' World!']
```

# Python - String Concatenation
* To concatenate, or combine, two strings you can use the + operator...

```python
#01. ExampleGet your own Python Server:- Merge variable a with variable b into variable c.

a = "Hello"
b = "World"
c = a + b
print(c)
```

```python
#02. Example:- To add a space between them, add a " ".

a = "Hello"
b = "World"
c = a + " " + b
print(c)
```

# Python - Format - Strings
* As we learned in the Python Variables chapter, we cannot combine strings and numbers like this...

```python
#01. Example:-

age = 36
#This will produce an error:
txt = "My name is John, I am " + age
print(txt)

#But we can combine strings and numbers by using f-strings or the format() method!
```

```python
#02. F-Strings:- F-String was introduced in Python 3.6, and is now the preferred way of formatting strings.

#To specify a string as an f-string, simply put an f in front of the string literal, and add curly brackets {} as placeholders for variables and other operations.

#Example:- Create an f-string.

age = 36
txt = f"My name is John, I am {age}"
print(txt)
```

```python
#03. Placeholders and Modifiers:- A placeholder can contain variables, operations, functions, and modifiers to format the value.

#Example:- Add a placeholder for the price variable:

price = 59
txt = f"The price is {price} dollars"
print(txt)
```

# Python - Escape Characters
* To insert characters that are illegal in a string, use an escape character.
* An escape character is a backslash \ followed by the character you want to insert.
* An example of an illegal character is a double quote inside a string that is surrounded by double quotes...

```python
#01. Example:- You will get an error if you use double quotes inside a string that is surrounded by double quotes.

txt = "We are the so-called "Vikings" from the north."
```

```python
#02. Example:- The escape character allows you to use double quotes when you normally would not be allowed:

txt = "We are the so-called \"Vikings\" from the north." # To fix this problem, use the escape character \".
```

```python
#03. Escape Characters

# Code	    Result	
# \'	 Single Quote	
# \\	 Backslash	
# \n	 New Line	
# \r	 Carriage Return	
# \t	 Tab	
# \b	 Backspace	
# \f	 Form Feed	
# \ooo	 Octal value	
# \xhh	 Hex value
```

# Python - String Methods
* Python has a set of built-in methods that you can use on strings.
* Note: All string methods return new values. They do not change the original string...

🔥 Very Important Methods.!!!

```python
#01. lower():- Converts a string into lower case.

txt = "PYTHON"
print(txt.lower())
```

```python
#02. upper():- Converts a string into upper case.

txt = "python"
print(txt.upper())
```

```python
#03. strip():- Extra spaces remove karta hai.

txt = "   hello   "
print(txt.strip())
```

```python
#04. split():- String ko parts me todta hai.

txt = "apple banana mango"
print(txt.split())
```

```python
#05. replace():- Text replace karta hai.

txt = "I like Java"
print(txt.replace("Java", "Python"))
```

```python
#06. find():- Character/word position find karta hai.

txt = "python"
print(txt.find("t"))
```

```python
#07. count():- Occurrences count karta hai.

txt = "apple apple"
print(txt.count("apple"))
```

```python
#08. startswith():- Returns true if the string starts with the specified value.

txt = "data.csv"
print(txt.startswith("data"))
```

```python
#9. endswith():- Returns true if the string ends with the specified value.

txt = "report.csv"
print(txt.endswith(".csv"))
```

```python
#10. join():- Strings combine karta hai.

words = ["I", "love", "Python"]
print(" ".join(words))
```

🔥 Data Validation Methods.!!!

```python
#11. isdigit():- Returns True if all characters in the string are digits.

txt = "123"
print(txt.isdigit())
```

```python
#12. isalpha():- Returns True if all characters in the string are in the alphabet.

txt = "Python"
print(txt.isalpha())
```

```python
#13. isalnum():- Returns True if all characters in the string are alphanumeric.

txt = "Python123"
print(txt.isalnum())
```

🔥 Formatting Methods.!!!

```python
#14. format():- Formats specified values in a string.

name = "Sufiyan"
print("My name is {}".format(name))
```

```python
#15. capitalize():- Converts the first character to upper case.

txt = "python"
print(txt.capitalize())
```

```python
#16. title():- Converts the first character of each word to upper case.

txt = "machine learning"
print(txt.title())
```

🔥 Useful in NLP / Text Cleaning.!!!

```python
#17. casefold():- Converts string into lower case.

txt = "HELLO"
print(txt.casefold())
```

```python
#18. splitlines():- Splits the string at line breaks and returns a list.

txt = "Hello\nWorld"
print(txt.splitlines())
```
