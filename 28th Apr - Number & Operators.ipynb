Logical Operators:- Used to combine boolean values.

```python
x = 10
```

```python
y = 20
```

```python
z = 5
```

```python
print(x > z and y > x) #AND
```

```python
print(x > y or y > z) #OR
```

```python
print(not(x > y)) #NOT
```

Unary Operators:- operators that operate on a single operand (one value).

```python
#Unary Plus (+)
x = 5
print(+x)   # Output: 5
```

```python
#Unary Minus (-)
x = 5
print(-x)   # Output: -5
```

```python
#Logical NOT (not)
x = True
print(not x)   # Output: False
```

```python
#Bitwise NOT (~)
x = 5          # Binary: 00000101
print(~x)      # Output: -6
```
