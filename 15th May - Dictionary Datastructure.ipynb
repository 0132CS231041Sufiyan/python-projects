Dictionary:- A dictionary is a built-in data type in Python used to store data in key-value pairs, & dictionary is mutable.

```python
#Syntax:
student = {
    "name": "Sufiyan",
    "age": 20,
    "course": "python"
}

print(student["name"])
print(student["course"])
```

```python
#01. keys():- Sabhi keys ko return karta hai.

student = {
    "name": "Sufiyan",
    "age": 20
}

print(student.keys())
```

```python
#02. values():- Sabhi values return karta hai.

student = {
    "name": "Sufiyan",
    "age": 20
}

print(student.values())
```

```python
#03. items():- Key-value pairs return karta hai.

student = {
    "name": "Sufiyan",
    "age": 20
}

print(student.items())
```

```python
#04. get():- Kisi key ki value safely access karta hai.

student = {
    "name": "Sufiyan",
    "age": 20
}

print(student.get("name"))
```

```python
#05. update():- Dictionary ko update karta hai.

student = {
    "name": "Sufiyan",
    "age": 20
}

student.update({"city": "Hyderabad"})

print(student)
```

```python
#06. pop():- Specific item remove karta hai.

student = {
    "name": "Sufiyan",
    "age": 20
}

student.pop("age")

print(student)
```

```python
#07. popitem():- Last inserted item remove karta hai.

student = {
    "name": "Sufiyan",
    "age": 20
}

student.popitem()
```

```python
#08. clear():- Pure dictionary ka data remove karta hai.

student = {
    "name": "Sufiyan",
    "age": 20
}

student.clear()

print(student)
```

```python
#09. copy():- Dictionary ki copy banata hai.

student = {
    "name": "Sufiyan",
    "age": 20
}

student.copy()

print(student)
```

```python
#10. setdefault():- Agar key exist nahi karti to new key add karta hai.

student = {
    "name": "Sufiyan",
    "age": 20
}

student.setdefault("course", "python")

print(student)
```

```python
#11. fromkeys():- Method ek new dictionary create karta hai using keys.

#Example 1

keys = ["name", "age", "city"]

d = dict.fromkeys(keys)

print(d)
```

```python
#Example 3

students = ["Ali", "Arif", "Sufiyan"]

marks = dict.fromkeys(students, 50)

print(marks)
```
