List:- A List in Python is a data structure used to store multiple items in a single variable, & lists are mutable, changeble...

```python
#0. Simple list
fruits = ["apple", "banana", "mango"]
fruits
```

```python
#01. append():- List ke end me ek item add karta hai.

fruits = ["apple", "banana"]

fruits.append("mango")

print(fruits)
```

```python
#02. insert():- Specific index pe value add karta hai.

numbers = [1, 2, 4]

numbers.insert(2, 3)

print(numbers)
```

```python
#03. extend():- Ek list ke saare items dusri list me add karta hai.

a = [1, 2]
b = [3, 4]

a.extend(b)

print(a)
```

```python
#04. remove():- Specific value delete karta hai.

fruits = ["apple", "banana", "mango"]

fruits.remove("banana")

print(fruits)
```

```python
#05. pop():- Index se value remove karta hai, by defoult remove last value...

numbers = [10, 20, 30]

numbers.pop(1)

print(numbers)
```

```python
#06. clear():- Puri list empty.

data = [1, 2, 3]

data.clear()

print(data)
```

```python
#07. index():- Item kis position pe hai, & first index count.

fruits = ["apple", "banana", "mango"]

print(fruits.index("banana"))
```

```python
#08. count():- Item kitni baar hai.

nums = [1, 2, 1, 1, 3]

print(nums.count(1))
```

```python
#09. sort():- Ascending order me sort.

nums = [5, 2, 8, 1]

nums.sort() #Ascending order & parameter tunning
#nums.sort(reverse = True) #Descending order & hyperparameter tunning

print(nums)
```

```python
#10. reverse():- List ko ulta karta hai.

nums = [1, 2, 3]

nums.reverse()

print(nums)
```

```python
#11. copy():- List ki duplicate copy.

a = [1, 2, 3]

b = a.copy()

print(b)
```

```python
#12. append vs extend.

a = [1, 2]

a.append([3, 4])
print(a)
```

```python
#13. extend():- Dusri list ke multiple items add karta hai.

a = [1, 2]

a.extend([3, 4])
print(a)
```
