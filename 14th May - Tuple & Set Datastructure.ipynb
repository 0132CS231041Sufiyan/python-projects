Tuple:- A tuple is an immutable data type in Python, which means its values cannot be changed after creation.
Because of this, tuples have only two built-in methods.

```python
#01. count():- Ye batata hai ki koi value tuple me kitni baar aayi hai.

t = (1, 2, 2, 3, 2)

print(t.count(2))
```

```python
#02. index():- Ye kisi value ka first index position return karta hai.

t = (10, 20, 30, 20)

print(t.index(20))
```

```python
#03. indexing & slicing(same as list).

t = (10, 20, 30, 40)

print(t[0])   #10
print(t[-1])  #40
print(t[1:3]) #(20, 30)
```

```python
#04. loops on tuples(sam as list).

t = (10, 20, 30, 40)

for i in t:
    print(i)
```

```python
#Q1. Using loops to calculate sum of all elements in the tuple.

t = (5, 15, 25)

sum = 0
for i in t:
    sum = sum + i
    
print("sum:", sum)
```

Set in Python:-A set in Python is a collection that.

1) Does not allow duplicate values
2) Is unordered
3) Is mutable (can be changed)

```python
#0. Creating a Set

numbers = {1, 2, 3, 4}

print(type(numbers))
```

```python
# Duplicate Values Are Removed Automatically.

data = {1, 2, 2, 3, 3, 4}

print(data)
```

```python
# Empty Set

# x = {}  #Wrong
x = set() #Correct
print(type(x))
```

```python
#01. add():- Adds a value to the set.

fruits = {"apple", "banana"}

fruits.add("mango")

print(fruits)
```

```python
#02. remove():- Removes a value from the set.

fruits = {"apple", "banana"}

fruits.remove("banana")

print(fruits)
```

```python
#03. discard():- Removes a value without giving an error if it does not exist.

a = {1, 2, 3}

a.discard(5)

print(a)
```

```python
#04. pop():- Removes a random item.

a = {10, 20, 30}

a.pop()

print(a)
```

```python
#05. clear():- Removes all items from the set.

a = {1, 2, 3}

a.clear()

print(a)
```

Set Operations:-1) Union (|), 2) Intersection (&), 3) Difference (-)...

```python
#01.Union (|):- Combines all values from both sets.

a = {1, 2}
b = {3, 4}

print(a | b)
```

```python
#02. Intersection (&):- Returns common values.

a = {1, 2, 3}
b = {2, 3, 4}

print(a & b)
```

```python
#03. Difference (-):- Returns unique values from the first set.

a = {1, 2, 3}
b = {2, 3}

print(a - b)
```
